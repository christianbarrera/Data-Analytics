# Job Outcomes
This project looks into the jobs outcomes of college graduates between 2010 and 2012.  Many factors are players in the outcomes, and some included in this project are gender diversity, employment rates, and salaries.  

## Questions to explore:  
Do students in more popular majors make more money?  
Do students that majored in subjects that were majority female make more money?

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
```

```python
recent_grads.plot(x='Sample_size', y='Median',
                 kind='scatter', title='Employed vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate',
                 kind='scatter', title='Employed vs. Unemployment_rate')
```

```python
recent_grads.plot(x='Full_time', y='Median',
                 kind='scatter', title='Full Time vs. Median')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate',
                 kind='scatter', title='Share of Women vs. Unemployment Rate')
```

```python
recent_grads.plot(x='Men', y='Median',
                 kind='scatter', title='Men vs. Median')
```

```python
recent_grads.plot(x='Women', y='Median',
                 kind='scatter', title='Women vs. Median')
```

### Do students in more popular majors make more money?  
In the scatter plots above, we see some answers to one of our questions.  The more popular majors do not bring in the most money, but do pay more than many of the most popular majors.  As can be seen in the employed vs unemployed scatter plot, the popular majors are about in the middle.  

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,20))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=25, title=str(cols[r]))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,14))
for r in range(5,8):
    ax = fig.add_subplot(3,1,r-4)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=25, title=str(cols[r]))
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(8,6))
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(8,6))
```

```python
ax = recent_grads[:10].plot.bar(x='Major', y='ShareWomen', title='Highest Paying',  legend=False)
ax.set_ylabel("Share of Women")
ax = recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', title='Lowest Paying', legend=False)
ax.set_ylabel("Share of Women")
```

### Do students that majored in subjects that were majority female make more money?
The above two bars graphs show a strong correlation that male dominated majors are the highest paying.  According to the data, the 10 highest paying majors have multiple shares of women hovering between 10% and 20%, while the 10 lowest paying majors have all shares of women above 60%.  Based on my observations around universities and corporations, tech companies and STEM programs are trying to promote gender equality and recruit more women to equalize these statistics.
